# PyCity Schools Analysis

In addition to comprehensive information about the city's schools, I also got to analize been the math and reading scores of every student. 

Turns out, most students did better in reading than in math. About 86% passed reading, but only 75% passed math.

The 'per_school_summary' breaks down how well each school did. Cabrera High School, a Charter school, did the best with a 91.3% passing rate. Rodriguez High School, a District school, did the worst with only 53% passing. Looking deeper, Charter schools generally have way better passing rates (over 35% higher) than District schools.

Now, talking about school size and performance, the 'size_summary' shows a kind of negative connection. Small and medium-sized schools have similar success levels. But in big schools, both average scores and passing rates drop a lot. Like, in small and medium-sized schools, 90% passed, but in big schools, only 58% did.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

```python
# Calculate the total number of unique schools
total_number_schools = len(school_data_complete["school_name"].unique())
total_number_schools
```

```python
# Calculate the total number of students
total_number_students = school_data_complete["Student ID"].count()
total_number_students
```

```python
# Calculate the total budget
total_budget = sum(school_data_complete['budget'].unique())
total_budget
```

```python
# Calculate the average (mean) math score
avg_math_score = round(school_data_complete['math_score'].mean(),6)
print(f"Average Math Score:  {avg_math_score}")
```

```python
# Calculate the average (mean) reading score
avg_reading_score = round(school_data_complete['reading_score'].mean(),5)
print(f"Average Reading Score:  {avg_reading_score}")
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
percentage_math_passing = round((school_data_complete[school_data_complete['math_score'] >= 70]['math_score'].count()/total_number_students)*100,6)
print(f"% Passing Math:  {percentage_math_passing}")
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
percentage_reading_passing = round((school_data_complete[school_data_complete['reading_score'] >= 70]['reading_score'].count()/total_number_students)*100,6)
print(f"% Passing Reading:  {percentage_reading_passing}")
```

```python
# Use the following to calculate the percentage of students that passed math and reading
overall_passing = round((school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >=70)]['student_name'].count()/total_number_students)*100,6)
overall_passing
print(f"% Overall Passing:  {overall_passing}")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools" : [total_number_schools],
     "Total Students" : [total_number_students],
     "Total Budget" : [total_budget],
     "Average Math Score" : [avg_math_score],
     "Average Reading Score" : [avg_reading_score],
     "% Passing Math" : [percentage_math_passing],
     "% Passing Reading" : [percentage_reading_passing],
     "% Overall Passing" : [overall_passing],
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## Scores by School Spending

```python
# Establish the bins 
bins = [0, 584.9, 629.9, 643.9, 675]
group_name = ['< $584', "$585 - 629", "$630 - 644", "$644 - 675"]
```

```python
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_name)


s_spending = school_data_complete.groupby('spending_bins')


avg_m = round(s_spending['math_score'].mean(),2)
avg_r = round(s_spending['reading_score'].mean(),2)
pass_m = round((school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/s_spending['Student ID'].count())*100,2)
pass_r = round((school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/s_spending['Student ID'].count())*100,2)
overall = round((school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/s_spending['Student ID'].count())*100,2)


scores_by_school_spending = pd.DataFrame({
    "Average Math Score": avg_m,
    "Average Reading Score": avg_r,
    "% Passing Math": pass_m,
    "% Passing Reading": pass_r,
    "% Overall Passing": overall
            
})
            


scores_by_school_spending.index.name = "Spending Ranges(Per Student)"
scores_by_school_spending = scores_by_school_spending.reindex(group_name)
scores_by_school_spending
```
